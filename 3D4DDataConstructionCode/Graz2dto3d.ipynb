```python
def assignCoordinates(number):
    if number+1==1:
        return [0,3]
    if number+1==2:
        return [1,1]
    if number+1==3:
        return [1,2]
    if number+1==4:
        return [1,3]
    if number+1==5:
        return [1,4]
    if number+1==6:
        return [1,5]
    if number+1==7:
        return [2,0]
    if number+1==8:
        return [2,1]
    if number+1==9:
        return [2,2]
    if number+1==10:
        return [2,3]
    if number+1==11:
        return [2,4]
    if number+1==12:
        return [2,5]
    if number+1==13:
        return [2,6]
    if number+1==14:
        return [3,1]
    if number+1==15:
        return [3,2]
    if number+1==16:
        return [3,3]
    if number+1==17:
        return [3,4]
    if number+1==18:
        return [3,5]
    if number+1==19:
        return [4,2]
    if number+1==20:
        return [4,3]
    if number+1==21:
        return [4,4]
    if number+1==22:
        return [5,3]
```

```python
import numpy as np
import scipy.io as sio

for file_path in file_list:
    mat_data = scipy.io.loadmat(file_path)
    shape = mat_data['data'].shape
    matrix = np.zeros((shape[0],6,7,shape[2]))
    print(shape)
    print(matrix.shape)
    for i in range(len(mat_data["data"])):
        for j in range(len(mat_data['data'][0])):
            for w in range (len(mat_data['data'][0][0])):
                matrix[i][assignCoordinates(j)[0]][assignCoordinates(j)[1]][w] = mat_data['data'][i][j][w]
    new_data = {'data': matrix, 'label': mat_data['label']}
    sio.savemat('3D'+ file_path.split("/")[-1], new_data)
```
