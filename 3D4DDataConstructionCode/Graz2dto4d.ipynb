```python
import glob
from google.colab import files
import scipy.io

# Get a list of .mat files in the current directory
file_list = glob.glob('/content/3dGrazNormalized/*T_P.mat')
```

```python
import numpy as np
import scipy.io as sio

for file_path in file_list:
    mat_data = scipy.io.loadmat(file_path)
    matrix = np.zeros((len(mat_data["data"]),6,7,4,1000))
    for i in range(len(mat_data["data"])):
        for j in range(len(mat_data['data'][0][0][0])):
            matrix[i][2][3][3][j] = mat_data['data'][i][2][3][j]

            matrix[i][1][3][2][j] = mat_data['data'][i][1][3][j]
            matrix[i][2][2][2][j] = mat_data['data'][i][2][2][j]
            matrix[i][3][3][2][j] = mat_data['data'][i][3][3][j]
            matrix[i][2][4][2][j] = mat_data['data'][i][2][4][j]

            matrix[i][0][3][1][j] = mat_data['data'][i][0][3][j]
            matrix[i][1][2][1][j] = mat_data['data'][i][1][2][j]
            matrix[i][2][1][1][j] = mat_data['data'][i][2][1][j]
            matrix[i][3][2][1][j] = mat_data['data'][i][3][2][j]
            matrix[i][4][3][1][j] = mat_data['data'][i][4][3][j]
            matrix[i][3][4][1][j] = mat_data['data'][i][3][4][j]
            matrix[i][2][5][1][j] = mat_data['data'][i][2][5][j]
            matrix[i][1][4][1][j] = mat_data['data'][i][1][4][j]

            matrix[i][1][1][0][j] = mat_data['data'][i][1][1][j]
            matrix[i][2][0][0][j] = mat_data['data'][i][2][0][j]
            matrix[i][3][1][0][j] = mat_data['data'][i][3][1][j]
            matrix[i][4][2][0][j] = mat_data['data'][i][4][2][j]
            matrix[i][5][3][0][j] = mat_data['data'][i][5][3][j]
            matrix[i][4][4][0][j] = mat_data['data'][i][4][4][j]
            matrix[i][3][5][0][j] = mat_data['data'][i][3][5][j]
            matrix[i][2][6][0][j] = mat_data['data'][i][2][6][j]
            matrix[i][1][5][0][j] = mat_data['data'][i][1][5][j]

    new_data = {'data': matrix, 'label': mat_data['label']}
    sio.savemat('4D'+  "_"+file_path.split("_")[-2]+ "_"+file_path.split("_")[-1], new_data)
```
