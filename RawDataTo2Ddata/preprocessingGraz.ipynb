```python
import scipy.io
import pandas as pd
import numpy as np
import mne
from scipy.io import savemat

a = scipy.io.loadmat('A01T.mat')
a_test = scipy.io.loadmat('A01E.mat')
a_test_label = scipy.io.loadmat('A01E_labels.mat')
```

```python
data = a['s']
data_test = a_test['s']
data_test_label = np.array(a_test_label['classlabel'])
```

```python
TYP = []
for i in range(len(a['h']['EVENT'][0][0]['TYP'][0][0])):
    TYP.append(a['h']['EVENT'][0][0]['TYP'][0][0][i][0])

POS = []
for i in range(len(a['h']['EVENT'][0][0]['POS'][0][0])):
    POS.append(a['h']['EVENT'][0][0]['POS'][0][0][i][0])
```

```python
TYP_test = []
for i in range(len(a_test['h']['EVENT'][0][0]['TYP'][0][0])):
    TYP_test.append(a_test['h']['EVENT'][0][0]['TYP'][0][0][i][0])

POS_test = []
for i in range(len(a_test['h']['EVENT'][0][0]['POS'][0][0])):
    POS_test.append(a_test['h']['EVENT'][0][0]['POS'][0][0][i][0])
```

```python
EVENT_CODE = [769, 770,771,772]

class1_pos = []
for i in range(len(TYP)):
        if TYP[i] == 769:
            if i != len(TYP)-1:
                class1_pos.append((POS[i],POS[i+1]))
            else:
                class1_pos.append((POS[i],len(data)-POS[i]-1))

class2_pos = []
for i in range(len(TYP)):
        if TYP[i] == 770:
            if i != len(TYP)-1:
                class2_pos.append((POS[i],POS[i+1]))
            else:
                class2_pos.append((POS[i],len(data)-POS[i]-1))
    
class3_pos = []
for i in range(len(TYP)):
        if TYP[i] == 771:
            if i != len(TYP)-1:
                class3_pos.append((POS[i],POS[i+1]))
            else:
                class3_pos.append((POS[i],len(data)-POS[i]-1))
    
class4_pos = []
for i in range(len(TYP)):
        if TYP[i] == 772:
            if i != len(TYP)-1:
                class4_pos.append((POS[i],POS[i+1]))
            else:
                class4_pos.append((POS[i],len(data)-POS[i]-1))

class_pos_test = []
for i in range(len(TYP_test)):
        if TYP_test[i] == 783:
            if i != len(TYP_test)-1:
                class_pos_test.append((POS_test[i],POS_test[i+1]))
            else:
                class_pos_test.append((POS_test[i],len(data_test)-POS_test[i]-1))

```

```python
class1 = []
class2 = []
class3 = []
class4 = []

for i in class1_pos:
    class1.append(data[i[0]:(i[0]+1000),:22].T)
class1.pop()
for i in class2_pos:
    class2.append(data[i[0]:(i[0]+1000),:22].T)

for i in class3_pos:
    class3.append(data[i[0]:(i[0]+1000),:22].T)

for i in class4_pos:
    class4.append(data[i[0]:(i[0]+1000),:22].T)
```

```python
for i in range(len(class_pos_test)):
    if data_test_label[i] == 1:
        class1.append(data_test[class_pos_test[i][0]:(class_pos_test[i][0]+1000),:22].T)
    if data_test_label[i] == 2:
        class2.append(data_test[class_pos_test[i][0]:(class_pos_test[i][0]+1000),:22].T)
    if data_test_label[i] == 3:
        class3.append(data_test[class_pos_test[i][0]:(class_pos_test[i][0]+1000),:22].T)
    if data_test_label[i] == 4:
        class4.append(data_test[class_pos_test[i][0]:(class_pos_test[i][0]+1000),:22].T)

```

```python
class1_train = class1[0:int(len(class1)*0.8)+1]
class1_test = class1[int(len(class1)*0.8)+1:]
class2_train = class2[0:int(len(class2)*0.8)+1]
class2_test = class2[int(len(class2)*0.8)+1:]
class3_train = class3[0:int(len(class3)*0.8)+1]
class3_test = class3[int(len(class3)*0.8)+1:]
class4_train = class4[0:int(len(class4)*0.8)+1]
class4_test = class4[int(len(class4)*0.8)+1:]
```

```python
trails_filtered = []

for i in class1_train:
    trails_filtered.append(mne.filter.filter_data(i,250, 4, 40))
for i in class2_train:
    trails_filtered.append(mne.filter.filter_data(i,250, 4, 40))
for i in class3_train:
    trails_filtered.append(mne.filter.filter_data(i,250, 4, 40))
for i in class4_train:
    trails_filtered.append(mne.filter.filter_data(i,250, 4, 40))
```

```python
trails_filtered_test = []

for i in class1_test:
    trails_filtered_test.append(mne.filter.filter_data(i,250, 4, 40))
for i in class2_test:
    trails_filtered_test.append(mne.filter.filter_data(i,250, 4, 40))
for i in class3_test:
    trails_filtered_test.append(mne.filter.filter_data(i,250, 4, 40))
for i in class4_test:
    trails_filtered_test.append(mne.filter.filter_data(i,250, 4, 40))
```

```python
arr_1 = np.ones(len(class1_train))
arr_2 = np.repeat(2, len(class2_train))
arr_3 = np.repeat(3, len(class3_train))
arr_4 = np.repeat(4, len(class4_train))
label = np.concatenate((arr_1, arr_2, arr_3, arr_4))

mdic = {"data": trails_filtered, "label": label}
savemat("A01T_P.mat", mdic)
```

```python
arr_1 = np.ones(len(class1_test))
arr_2 = np.repeat(2, len(class2_test))
arr_3 = np.repeat(3, len(class3_test))
arr_4 = np.repeat(4, len(class4_test))
label = np.concatenate((arr_1, arr_2, arr_3, arr_4))

mdic = {"data": trails_filtered_test, "label": label}
savemat("A01E_P.mat", mdic)
```
