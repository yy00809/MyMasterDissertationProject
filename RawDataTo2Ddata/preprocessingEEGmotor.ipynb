```python
import os
import mne
import pyedflib
import scipy.io as sio
import numpy as np
# Specify the path of the directory
directory_path = '/content/extracted_folder/files/S001'

for i in range(105, 110):  # Change the range as needed
    directory_path = directory_path[:-3] + '{:03d}'.format(i)

    file_name = 'S'+'{:03d}'.format(i)+'R03.edf'
    print(file_name)
    for _ in range(11):
        name, extension = os.path.splitext(file_name)
        file_path = os.path.join(directory_path, file_name)
        data = mne.io.read_raw_edf(file_path)
        raw_data = data.get_data()
        data = mne.filter.filter_data(raw_data,160, 4, 40)
        data = data.T
        file = pyedflib.EdfReader(file_path)  
        annotations = file.readAnnotations()
        file.close()

        if 'R03' in file_name or 'R04' in file_name or 'R07' in file_name or 'R08' in file_name or 'R11' in file_name or 'R12' in file_name:
            for i in range(len(annotations[0])):
                if annotations[2][i] == 'T1':
                    left_fist.append(data[int(annotations[0][i]*160):int((4+annotations[0][i])*160)][:].T)
                if annotations[2][i] == 'T2':
                    right_fist.append(data[int(annotations[0][i]*160):int((4+annotations[0][i])*160)][:].T)
        if 'R05' in file_name or 'R06' in file_name or 'R09' in file_name or 'R10' in file_name or 'R13' in file_name or 'R14' in file_name:
            for i in range(len(annotations[0])):
                if annotations[2][i] == 'T1':
                    both_fists.append(data[int(annotations[0][i]*160):int((4+annotations[0][i])*160)][:].T)
                if annotations[2][i] == 'T2':
                    both_feet.append(data[int(annotations[0][i]*160):int((4+annotations[0][i])*160)][:].T)
    
        
        numeric_part = name[4:]

        # Increment the numeric part
        new_numeric_part = int(numeric_part[1:]) + 1

        file_name =  name[:4] + 'R{:02d}'.format(new_numeric_part) + extension

    subject_data = {'data': [], 'label': []}

    for j in range(len(left_fist)):
        subject_data['data'].append(left_fist[j])
        subject_data['label'].append(1)
    for j in range(len(right_fist)):
        subject_data['data'].append(right_fist[j])
        subject_data['label'].append(2)
    for j in range(len(both_fists)):
        subject_data['data'].append(both_fists[j])
        subject_data['label'].append(3)
    for j in range(len(both_feet)):
        subject_data['data'].append(both_feet[j])
        subject_data['label'].append(4)
   
    sio.savemat(directory_path[-4:]+'.mat', subject_data)

    left_fist = []
    right_fist = []
    both_fists = []
    both_feet = []
```
